## Multiline Plots
+ It is possible to visualize multiple plots in the same output
+ Matplotlib automatically assigns colors to the curves separately.

```python
x = np.arange(7)
plt.plot(x, -x**2)
plt.plot(x, -x**3)
plt.plot(x, -2*x)
plt.plot(x, -2**x)
plt.show()
```

You can write the same code in a simple way as follows:

```python
plt.plot(x, -x**2, 
        x, -x**3,
        x, -2*x, 
        x, -2**x)
plt.show()
```

## Grid, Axes, and Labels

```python
x = np.arange(3)
plt.plot(x, -x**2, 
        x, -x**3, 
        x, -2*x, 
        x, -2**x)
plt.grid(True)
plt.savefig('test.png')
plt.show()
```

You can see that the limits of the axes are set by default as follows

```python
x = np.arange(3)
plt.plot(x, -x**2, 
        x, -x**3, 
        x, -2*x, 
        x, -2**x)
plt.grid(True)
print(plt.axis())
plt.show()
```

You can also customize the values of the axes as follows:

```python
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
plt.grid(True)
plt.axis([0, 2, -8, 0])
print(plt.axis())
plt.show()
```

```python
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
plt.grid(True)
plt.xlim([-1,3])
plt.ylim([-10,0])
print(plt.axis())
plt.show()
```

```python
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
plt.grid(True)
plt.xlim([-0.5,2.5])
plt.ylim([-9,1])
plt.title('Simple Plot Demo')
plt.ylabel('y=f(x)')
plt.xlabel('x=np.arange(3)')
plt.show()
```

```python
x = np.arange(3)
plt.plot(x, -x**2, label='-x**2')
plt.plot(x, -x**3, label='-x**3')
plt.plot(x, -2*x, label='-2*x')
plt.plot(x, -2**x, label='-2**x')
plt.legend()

plt.grid(True)
plt.xlabel('x=np.arange(3)')
plt.xlim([0,2])

plt.ylabel('y=f(x)')
plt.ylim([-8,0])

plt.title('Simple plot demo')
plt.show()
```

```python
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
plt.legend(['-x**2', '-x**3', '-2*x', '-2**x'])
plt.grid(True)
plt.xlim([-0.5,2.5])
plt.ylim([-9,1])
plt.title('Simple Plot Demo')
plt.ylabel('y=f(x)')
plt.xlabel('x=np.arange(3)')
plt.show()
```

Change the location of the legend

```python
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
plt.legend(['-x**2', '-x**3', '-2*x', '-2**x'],
          loc='lower center')
plt.grid(True)
plt.xlim([-0.5,2.5])
plt.ylim([-9,1])
plt.title('Simple Plot Demo')
plt.ylabel('y=f(x)')
plt.xlabel('x=np.arange(3)')
plt.show()
```

save the visualization to disk with the following code

```python
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
plt.legend(['-x**2', '-x**3', '-2*x', '-2**x'],
          loc='lower center')
plt.grid(True)
plt.xlim([-0.5,2.5])
plt.ylim([-9,1])
plt.title('Simple Plot Demo')
plt.ylabel('y=f(x)')
plt.xlabel('x=np.arange(3)')
plt.savefig('test.png')
plt.show()
```

```python
plt.plot(x, y+0.4, 'g', x, y+0.2, 'y', x, y, 'r',x, y - 0.2, 'c',
         x, y - 0.4, 'k', x, y - 0.6, 'm', x, y - 0.8, 'w',
         x, y - 1, 'b')
plt.show()
```

## Object-Oriented Plotting
+ we are using the axis object to plot and set the labels and a title.

```python
x = np.arange(3)
fig, ax = plt.subplots()

ax.plot(x, -x**2, label='-x**2')
ax.plot(x, -x**3, label='-x**3')
ax.plot(x, -2*x, label='-2*x')
ax.plot(x, -2**x, label='-2**x')

ax.set_xlabel('x = np.arange(3)')
ax.set_ylabel('y = f(x)')
ax.set_title('Simple Plot Demo')

ax.legend()
ax.grid(True)
plt.show()
```

You can also add the text with the functions `ax.text()` or the function `plt.text()`.

```python
x = np.arange(3)
fig, ax = plt.subplots()

ax.plot(x, -x**2, label='-x**2')
ax.plot(x, -x**3, label='-x**3')
ax.plot(x, -2*x, label='-2*x')
ax.plot(x, -2**x, label='-2**x')

ax.set_xlabel('x = np.arange(3)')
ax.set_ylabel('y = f(x)')
ax.set_title('Simple Plot Demo')

ax.legend()
ax.grid(True)

ax.text(0.25, -5, 'Simple plot demo')

plt.show()
```

## Subplots
+ You can show multiple separate graphs in the same output.
+ Subplots can have their own titles, own labels, and other specifications
+ Subplots are created in a grid. 
+ The first subplot position is at the top left. 
+ The other subplot positions are relative to the first position.

```python
x = np.arange(3)
plt.subplots_adjust(wspace=0.3,
                   hspace=0.3)

plt.subplot(2,2,1)
plt.plot(x, -x**2)

plt.subplot(2,2,2)
plt.plot(x, -x**3)

plt.subplot(2,2,3)
plt.plot(x, -2*x)

plt.subplot(2,2,4)
plt.plot(x, -2**x)

plt.show()
```

You can write the same code in object-oriented fashion as follows:

```python
x = np.arange(3)

fig, axs = plt.subplots(2,2)
plt.subplots_adjust(wspace=0.3, hspace=0.3)

axs[0,0].plot(x, -x**2)
axs[0,1].plot(x, -x**3)
axs[1,0].plot(x, -2*x)
axs[1,1].plot(x, -2**x)

plt.show()
```
